```python
class Human:
    ## define some variables (attributes)
    ## init method( constructor )

    lifespan = 70 # class variable
    hands = 2 # class variables
    legs = 2
    eyes = 2
    
    def __init__(self, name, gender = None, friend_list = None ):
        '''
        this function will be executed when the class is instantiated
        '''
        print(" Human is being created!! ")
        self.name = name  # self indicates the instance in question
        self.gender = gender
        self.friend_list = friend_list if friend_list else []

    ## define some methods

    def add_friend(self, human):
        self.friend_list.append(human)
        human.friend_list.append(self)
        print(f"{self.name} and {human.name} are now friends!!! ")

    def __str__(self): ## what to return when the instance is printed ( string form)
        return f"{self.name} - {self.gender} "

    def __repr__(self): ## representation of the instance
        # return f"Name: {self.name} | Gender: {self.gender} "
        return f"{self.name}({self.gender}) "
        
    def walk(self): # instance method --> it is taking the instance as a parameter (through self)
        print("I am walking!!!")

    def eat(self): # instance method --> it is taking the instance as a parameter (through self)
        print("Hello I am eating!!!")

    # write a method that will change the lifespan of humans
    
    @classmethod  # this will indicate that the first parameter is a class
    def change_lifespan(cls, new_lifespan):
        '''
        This will change the lifespan for the entire class
        '''
        cls.lifespan = new_lifespan

    def change_lifespan_for_instance(self, new_lifespan):
        '''
        This is an instance method. this will change the lifespan for an instance
        '''
        self.lifespan = new_lifespan
        
```

## Abstraction
    - Interface /Promise 
    - Implementation

    - We want to enforce a certain interface. We want all payment methods to have a interface / promise of pay and cancel. 
    - Implementation of pay and cancel can be anything, but the interface should be present. Interface should be guranteed. 

    Abstract method: ---> is a method that doesnt have implementation ( not a concrete method)

```python
class CreditCardPayment:
    def pay(self):
        print("Transaction complete!!!")

    def cancel(self):
        print("Transaction cancelled!!!")

class NetBankingPayment:
    def make_payment(self):
        print("Transaction complete!!!")

    def cancel_payment(self):
        print("Transaction cancelled!!!")

class UPIPayment:
    def pay(self):
        print("Transaction complete!!!")

    def cancel(self):
        print("Transaction cancelled!!!")



```

```python
cc_payment = CreditCardPayment()
upi_payment= UPIPayment()
net_banking_payment = NetBankingPayment()


for payment_method in [cc_payment, upi_payment, net_banking_payment]:
    payment_method.pay()
```

```python
from abc import abstractmethod , ABC

'''
This code defines what qualifies as a PaymentMethod. This will ensure/enforce that all
payment methods will mandatorily have pay and cancel methods. 
(Mandatory method overriding for all abstract methods)
'''

class PaymentMethod(ABC):
    '''
    Interface
    '''
    @abstractmethod
    def pay(self):
        pass

    @abstractmethod
    def cancel(self):
        pass


class CreditCardPayment(PaymentMethod):
    def pay(self):
        '''
        implementation
        '''
        print("Transaction complete!!!")

    def cancel(self):
        print("Transaction cancelled!!!")

class NetBankingPayment(PaymentMethod):
    def make_payment(self):
        print("Transaction complete!!!")

    def cancel_payment(self):
        print("Transaction cancelled!!!")

class UPIPayment(PaymentMethod):
    def pay(self):
        print("Transaction complete!!!")

    def cancel(self):
        print("Transaction cancelled!!!")

```

```python
class BankAccount:

    
    def __init__(self, account_number,  holder_name, balance, branch_name = None,):
        self.account_number = account_number
        self.holder_name = holder_name
        # adding a underscore indicated that this is a protected attribute
        self.__balance = balance # this is protected/private attribute. we need to control the access to this !!
        self.branch_name = branch_name


    def deposit(self, amount: int):
        '''
        Needed to add some balance
        '''
        if amount < 0 :
            raise Exception("INVALID AMOUNT ENTERED!")
        self.__balance = self.__balance + amount
        print(f"{amount} deposited! Your new balance is {self.__balance}!!")

    def withdraw(self, amount):
        '''
        Needded to withdraw some money
        '''
        if amount < 0 :
            raise Exception("INVALID AMOUNT ENTERED!")
        if amount > self.__balance:
            raise Exception("INSUFFICIENT FUNDS!!")
        self.__balance = self.__balance - amount
        print(f"{amount} withdrawn! Your new balance is {self.__balance}!!")
        
    
```

## Abstraction Example

```python
class Cricketer(ABC):  # this is just an interface!!!!!!!!

    '''
    This is just interface. There is no implementation
    ## Cricketer class is abstract class. It CANNOT be instantiated !
    '''
    
    @abstractmethod
    def bat(self):
        '''
        This has to be mandatory overwritten!!!
        '''
        pass

    @abstractmethod
    def bowl(self):
        '''
        This has to be mandatory overwritten!!!
        '''
        pass

    @abstractmethod
    def field(self):
        '''
        This has to be mandatory overwritten!!!
        '''
        pass

    def talk(self):  # what is the need for this?
        '''
        this is a concrete method. May or may not be overridden!!
        '''
        print("Hello i am talking as a cricketer. ")
```

```python
cricketer = Cricketer()
```

```python
## An abstract class CANNOT be instantiated. We can CANNOT create a object from a abstract class. 
 ( There is no need to create object of abstract class. Because it has only abstract methods. )
```

```python
class IPLCricketer(Cricketer): # it means that IPLCricketer "is a" Cricketer. The definition or mandatory conditions to be fulfilled to qualify as a cricketer of Cricketer

    def bat(self):
        '''
        this is a concrete method. We need to the implementation here
        '''
        print("I am batting very fast. Sometimes pinch hitting")

    def bowl(self):
        print("I am struggling to bowl in a IPL match")

    def field(self):
        print("I am trying to save runs in a IPL match")
    

class TestCricketers(Cricketer):

    def bat(self):
        print("I am batting in a test match!!")

    def bowl(self):
        print("I am bowling in a test match!!")

    def field(self):
        print("I am fielding in a test match!!")
```

## @property decorator
    - this allows us to access a methods like an attribute while providing access control to get and set the attribute. 
    
    - getter,  setter, and deleter

    

```python
class BankAccount:

    def __init__(self, balance):
        self.balance = balance
    
    def deposit():
        pass

    def withdraw():
        pass

        
```

```python
class BankAccount:

    def __init__(self, balance):
        self._balance = balance
        pass
        # self.balance = balance

    @property  # now this balance is a property
    def balance(self):  # i want this balance to be accessed like an attribute
        ## this is a getter. This helps us get the value of a property
        '''
        here  i can add some custom logic
        
        '''
        print("you are trying to access your balance!!!")
        if not self._balance:
            return "there is no balance"
        return self._balance

    @balance.deleter
    def balance(self):
        
        self._balance = None
        print("balance is now deleted!!!!!")

        
    @balance.setter
    def balance(self, val):
        password = input ("Enter the password!!!")
        if password == "123":
            
            print("you are trying to set some value")
            if val > 0 :
                self._balance = val
            else:
                raise Exception("INVALID ACTION!!")
        else:
            raise Exception("INVALID PASSWORD!!")
        
     
    
    def deposit():
        pass

    def withdraw():
        pass
```
