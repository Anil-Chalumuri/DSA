# Two Pointer Approach

### 11. Container With Most Water

You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.



````python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        '''

        initialise a left and right pointer to both ends
        calculate the area = (right - left) * min(height[left], height[right])

        then move the left or right whichever is lower
        calculate the area again and keep the higher area
        '''
        max_area = 0
        left = 0
        right = len(height) - 1

        while left < right:
            area = (right - left) * min(height[left], height[right])
            max_area = max(area, max_area)
            # if area > max_area:
            #     max_area = area

            ## we will move either the left or right pointer
            if height[left] < height[right]:
                left = left + 1
            else:
                right = right - 1

        return max_area
````

### 977. Squares of a Sorted Array

```Example 1:

Output: [0,1,9,16,100]
Explanation: After squaring, the array becomes [16,1,0,9,100].
After sorting, it becomes [0,1,9,16,100].
Example 2:

Input: nums = [-7,-3,2,3,11]
Output: [4,9,9,49,121]
```


````python
class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:

        res_array = []
        left =0
        right = len(nums) - 1

        while left <= right:
            if nums[right]**2 > nums[left]**2:
                res_array.insert(0, nums[right]**2) ## this is O(n) operation. Bad Practice !! We can append and then reverse!!
                right = right - 1

            else:
                res_array.insert(0, nums[left] ** 2)
                left = left + 1

        return res_array


class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:

        res_array = []
        left =0
        right = len(nums) - 1

        while left <= right:
            if nums[right]**2 > nums[left]**2:
                res_array.append( nums[right]**2)
                right = right - 1

            else:
                res_array.append(nums[left] ** 2)
                left = left + 1

        return list(reversed(res_array))

````

### 567. Permutation in String

Given two strings s1 and s2, return true if s2 contains a permutation of s1, or false otherwise.

In other words, return true if one of s1's permutations is the substring of s2.


```
Example 1:

Input: s1 = "ab", s2 = "eidbaooo"
Output: true
Explanation: s2 contains one permutation of s1 ("ba").
Example 2:

Input: s1 = "ab", s2 = "eidboaoo"
Output: false

```

Constraints:

1 <= s1.length, s2.length <= 104
s1 and s2 consist of lowercase English letters.

````python
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        '''
        Step 1: Create the target counter for s1 and curr_counter
        Step 2: Fixed window in s2 of size len(s1)
        In loop till we reach end of s2
            Step 3: Add the characters in curr_counter as per the window
                    ( increase the entering element and decrease the leaving element)

            Step 4: Compare the target_counter and curr_counter

        Time complexity : O(m + n) # m is s1 and n is s2
        Space Complexity : O(26) # constant space

        '''

        # edge case handled
        if len(s2) < len(s1):
            return False

        target_counter, curr_counter = defaultdict(int), defaultdict(int)

        for i in range(len(s1)): # it will run n times ( n = len(s1))
            target_counter[s1[i]] += 1
            curr_counter[s2[i]] += 1

        if target_counter == curr_counter:
            return True

        for i in range(len(s1), len(s2)):
            entering_element = s2[i]
            leaving_element = s2[i-len(s1)]

            # updating the curr_counter as per the current window
            curr_counter[entering_element] += 1
            curr_counter[leaving_element] -= 1
            if curr_counter[leaving_element] == 0:
                del curr_counter[leaving_element]
a
            ## compare the curr_counter and target_counter
            if curr_counter == target_counter :  ## max 26 comparisons
                return True

        return False
````
