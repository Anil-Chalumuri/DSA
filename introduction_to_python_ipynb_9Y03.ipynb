````python
# return the position of the even numbers in the list

nums = [2, 43, 10, 101, 32]  
result = []

for index in range(len(nums)):
    num = nums[index]
    if num % 2 == 0: ## condition to check if the number is even
        result.append(index)

print(result)
        
````

````python
# return the position of the even numbers in the list

nums = [2, 43, 10, 101, 32]  
result = []

for index in range(len(nums)):
    num = nums[index]
    if num % 2 == 0: ## condition to check if the number is even
        result.append((index, nums[index]))

print(result)
        
````

````python
# find out if any element in the list is odd

nums = [10, 12, 13, 14, 100]

for num in nums:
    print(num)
    if num%2 != 0:  # condition for odd number
        print(f"{num} is odd")
        break
        
````

````python
## write a function to print the table of a number

def print_table(num, count = 10):
    for i in range(1, count+1):
        print(f"{num} * {i} = {num*i} ")

print_table(10, 20)  # positional argument

print_table(num = 10, count = 20)  # keyword argument
````

````python
print_table(10, count = 20)  # keyword argument
````

````python
print_table(num = 10,  20)  # keyword argument
````
